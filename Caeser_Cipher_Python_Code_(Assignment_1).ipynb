```python
import string
letters=string.printable
print(letters)
```

Encrpytion

```python
dict={}
key=input('the secret key=')
key=int(key)
for i in range (len(letters)):
  dict[letters[i]]=letters[(i+key)%len(letters)]
print (dict)
```

```python
p=input('Enter plaintext = ')
ciphertxt=[]
for char in p:
  temp=dict[char]
  ciphertxt.append(temp)

ciphertxt="".join(ciphertxt)
print('cipher text is =',ciphertxt)
```
